```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
url = 'http://books.toscrape.com/catalogue/page-6.html'
html=requests.get(url).content
print(html)
```

```python
soup=BeautifulSoup(html,'lxml')
print(soup)
```

```python
# Es para seleccionar un tipo de elemento en especifico
soup_2=soup.select('h3>a')
```

```python
titulo=[element['title'] for element in soup_2]
titulo
```
